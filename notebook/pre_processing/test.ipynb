```python
# Import lại thư viện sau khi môi trường bị reset
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Dữ liệu mẫu
y_real = np.array([3, 5, 7, 9, 11])
y_pred = np.array([2.8, 5.1, 6.9, 9.2, 10.8])

# Tính hệ số tương quan R
R, _ = pearsonr(y_real, y_pred)

# Vẽ đồ thị mô phỏng hệ số tương quan
plt.figure(figsize=(6, 5))
sns.regplot(x=y_real, y=y_pred, scatter_kws={'s':50}, line_kws={'color':'red'})
plt.xlabel("Giá trị thực tế (y)")
plt.ylabel("Giá trị dự đoán (ŷ)")
plt.title(f"Minh họa hệ số tương quan R = {R:.4f}")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Tạo dữ liệu giả lập
np.random.seed(42)
X = np.linspace(0, 10, 100)  # Dữ liệu X từ 0 đến 10
y = 2 * X + 1 + np.random.normal(0, 2, 100)  # Y có nhiễu

X = X.reshape(-1, 1)  # Chuyển đổi X thành ma trận cột

# Huấn luyện mô hình hồi quy
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Vẽ Confidence Band
plt.figure(figsize=(8, 5))
sns.regplot(x=X.flatten(), y=y, ci=95, line_kws={'color': 'red'})  # Đường hồi quy màu đỏ
plt.xlabel("Diện tích (m²)")
plt.ylabel("Giá nhà (triệu VND)")
plt.title("Đường hồi quy và Confidence Band")
plt.show()
```
