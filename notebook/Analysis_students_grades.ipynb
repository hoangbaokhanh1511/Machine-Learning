Analysis of student grades

```python
import numpy as np
import matplotlib as plt
```

1. Score Data Generation: Tạo 1 mảng dữ liệu Numpy ngẫu nhiên chứa điểm số của học sinh (từ 0 đến 100) cho 5 môn học với 300 học sinh.

```python
np.random.seed(0)
b = np.random.randint(0,100, (300,5)) 
# Học sinh tượng trưng cho các hàng (rows) trong ma trận và các môn tượng trưng cho các cột (column) 
# trong ma trận.
print("SCORES OF EACH STUDENT: ")
print(b)
```

```python
# Subject 1:
print(b[:,0])
# subject 2:
print(b[:,1])
```

2. Data Analysis 

a. Calculate the GPA of each student

```python
average_each_student = np.mean(b, axis = 1) 
print("AVERAGE SCORE OF EACH STUDENT: ")
print(average_each_student)
```

b. Calculate the average score of each subject

```python
average_each_subject = np.mean(b, axis= 0)
print("AVERAGE SCORE OF EACH SUBJECT:")
print(average_each_subject)
```

c. Identify students with highest and lowest GPAs:

```python
student_with_highest_GPAs = np.argmax(average_each_student)
score_of_the_highest_student = [b[i] for i in range(len(average_each_student)) if i == student_with_highest_GPAs]
```

```python
print("STUDENT WITH THE HIGHEST GPAs IN 300 STUDENTS OF THE SCHOOL:", student_with_highest_GPAs)
print("THE AVERAGE SCORE OF THE STUDENT IS:", score_of_the_highest_student)
```

```python
student_with_lowest_GPAs = np.argmin(average_each_student)
score_of_the_lowest_student = [b[i] for i in range(len(average_each_student)) if i == student_with_lowest_GPAs]
```

```python
print("STUDENT WITH THE LOWEST GPAs IN 300 STUDENTS OF THE SCHOOL:", student_with_lowest_GPAs)
print("THE AVERAGE SCORE OF THE STUDENT IS:", score_of_the_lowest_student)
```

d. The most difficult subject:

```python
difficult_subject = np.argmin(average_each_subject)
print("THE MOST DIFFICULT SUBJECT IS: SUBJECT", difficult_subject + 1)
```

e. Classify students based on GPA (A, B, C, D): 


```python
list_classification = []
for i in average_each_student:
    if 75 < i < 100: 
        list_classification.append('A')
    elif 50 < i:
        list_classification.append('B')
    elif 25 < i:
        list_classification.append('C')
    else:
        list_classification.append('D')
```

```python
print("LIST OF CLASSIFICATION OF STUDENTS BY GPA: ")
print(list_classification)
```

f. Count the number of students in each category:

```python
def count_list(input_list, x):
    count = 0
    for i in input_list:
        if i == x: 
            count = count + 1
    return count
print("Số lượng A trong danh sách:", count_list(list_classification, 'A'))
print("Số lượng B trong danh sách:", count_list(list_classification, 'B'))
print("Số lượng C trong danh sách:", count_list(list_classification, 'C'))
print("Số lượng D trong danh sách:", count_list(list_classification, 'D'))
```

g. Determine which students have the highest and lowest scores in subject 1 (tương tự cho 4 cái còn lại)

```python
list_subject1 = b[:, 0]
print("List scores of the subject1: ")
print(list_subject1)
```

```python
# Highest and lowest scores in subject 1:
students_with_highest_score_subject1 = [i for i in range(len(list_subject1)) if list_subject1[i] == max(list_subject1)] 
students_with_lowest_score_subject1 = [i for i in range(len(list_subject1)) if list_subject1[i] == min(list_subject1)]
```

```python
print("All the students have the best score in subject1 is ", students_with_highest_score_subject1)
print("The highest score is: ", max(list_subject1))
print("All the students have the lowest score in subject1 is ", students_with_lowest_score_subject1)
print("The highest score is: ", min(list_subject1))
```

1. Sử dụng biểu đồ Scatter để biểu thị mối tương quan giữa điểm thi môn 1 và môn 2:


```python
import matplotlib.pyplot as plt
```

```python
list_subject2 = b[:,1]
plt.figure(figsize=(8,6))
plt.scatter(range(len(list_subject1)), list_subject1, c= 'red', alpha= 0.7, label = 'Môn 1' )
plt.scatter(range(len(list_subject2)), list_subject2, c = 'blue', alpha= 0.7, label = 'Môn 2')
plt.grid(True)
plt.xlabel('Học sinh')
plt.ylabel('Điểm số')
plt.legend()
plt.title("Mối tương quan giữa giữa điểm môn 1 và điểm môn 2");
```

2. Draw a histogram chart comparing the score of each subject (Subject 1,2,3,4):

```python
list_subject3 = b[:,2]
list_subject4 = b[:,3]
```

```python
fig, axs = plt.subplots( 2,2, figsize = (15,15))
axs[0][0].hist(list_subject1,bins=30, color = 'green', edgecolor= 'black')
axs[0][0].set_title("Điểm môn 1")
axs[0][0].set_xticks(np.arange(0, 101, 5))  # Hiển thị các số từ 0 đến 100 với bước nhảy 5
axs[0][0].grid(axis='y', linestyle='--', alpha=0.7)  # Thêm lưới dọc

axs[0][1].hist(list_subject2, bins = 30, color = 'blue', edgecolor = 'black')
axs[0][1].set_title("Điểm môn 2")
axs[0][1].set_xticks(np.arange(0, 101, 5)) 
axs[0][1].grid(axis='y', linestyle='--', alpha=0.7) 

axs[1][0].hist(list_subject3, bins = 30, color = 'violet', edgecolor = 'black')
axs[1][0].set_title("Điểm môn 3")
axs[1][0].set_xticks(np.arange(0, 101, 5)) 
axs[1][0].grid(axis='y', linestyle='--', alpha=0.7) 

axs[1][1].hist(list_subject4, bins = 30, color = 'red', edgecolor = 'black')
axs[1][1].set_title("Điểm môn 4")
axs[1][1].set_xticks(np.arange(0, 101, 5)) 
axs[1][1].grid(axis='y', linestyle='--', alpha=0.7) 
plt.tight_layout()  # Sắp xếp các biểu đồ gọn gàng
```
