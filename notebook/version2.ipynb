Data Visualization - California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# import seaborn 
```

```python
cities = pd.read_csv('F:\DataMining\dataset\california_cities.csv')
cities.head()
```

```python
# Extract Latd = Vĩ độ, Longd = Kinh độ
lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']

plt.style.use('seaborn-v0_8-darkgrid',)
plt.figure(figsize=(8,6))
#Plot ussing Pyplot API
plt.scatter(lat, lon,
            c = np.log10(population), cmap='inferno', s = area,
            linewidths=0, alpha= 0.5)

plt.axis('equal');
plt.xlabel('Lattitude')
plt.ylabel('Longtitude')
plt.colorbar(label = 'log$_{10}$(population)')
plt.clim(3,7)

#create a legend for cities's sizes 
area_range = [50, 100, 300, 500]
for area in area_range:
    plt.scatter([], [], s=area, c = 'k', alpha= 0.4,
                label = str(area) + 'km$^2$')
plt.legend(labelspacing = 1, title = 'City Area')
plt.title('California Cities: Population and Area Distribution');
```
