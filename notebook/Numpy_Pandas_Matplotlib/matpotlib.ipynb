Thư viện Matplotlib trong python

Để tạo một đồ thị đơn giản với Matpotlib, bạn bắt đầu bằng cách tạo một Figure, sau đó thêm Axes vào Figure, và cuối cùng là vẽ trên Axes. Dưới đây là ví dụ thực hiện:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create data:
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots(figsize=(8, 6)) 


# Vẽ dữ liệu trên Axes:
ax.plot(x, y, label = 'sin(x)')

# tuỳ chỉnh trên đồ thị: 
ax.set_title('SIN WAVE') # Tiêu đề của đồ thị
ax.set_xlabel('X Axis') # Nhãn cho trục X
ax.set_ylabel('Y Axis') # Nhãn cho trục Y
ax.legend() # Hiển thị huyền thoại
plt.show()
```

<!-- 1. Vẽ đồ thị đường: Đây là đồ thị phổ biến nhất, thường được sử dụng để hiển thị sự thay đổi của dữ liệu qua thời gian. -->

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, color='blue', linestyle='-', linewidth=2, marker='o', label='sin(x)')
plt.title('Đồ Thị Đường Sin(x)')
plt.xlabel('X')
plt.ylabel('Sin(X)')
plt.legend()
plt.show()
```

2. Vẽ đồ thị cột

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 5]
plt.bar(categories, values, color = 'green')
plt.title('Đồ thị cột')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 5]
fig, ax = plt.subplots()
ax.bar(categories, values, color = 'b')
ax.set(title='Đồ thị cột', xlabel='Categories', ylabel='Values');
```

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 5]
fig, ax = plt.subplots()
ax.barh(categories, values, color = 'b')
ax.set(title='Đồ thị cột', xlabel='Categories', ylabel='Values');
```

3. Vẽ đồ thị Scatter: Đồ thị Scatter được dùng để hiển thị mối quan hệ giữa 2 biến

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y, color = 'red')
plt.title('Đồ thị Scatter')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000*rng.rand(100)
fig, ax = plt.subplots()
ax.scatter(x, y, s=sizes, c = colors, alpha = 0.8);
```

```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.RandomState(0)

# Tạo 1000 phần tử ngẫu nhiên
x = rng.randn(1000)
y = rng.randn(1000)
colors = rng.rand(1000)
sizes = 1000 * rng.rand(1000)

# Vẽ biểu đồ scatter
fig, ax = plt.subplots()
ax.scatter(x, y, s=sizes, c=colors, alpha=0.8)

# Hiển thị biểu đồ
plt.show()
```

4. Vẽ biểu đồ hình tròn: Biểu đồ hình tròn được sử dụng để hiển thị tỷ lệ phần trăm của mỗi phần so với tổng thể

```python
sizes = [10, 20, 30, 40]
labels = ['A', 'B', 'C', 'D']

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Biểu Đồ Hình Tròn')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, alpha=0.5, color='blue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Giá Trị')
plt.ylabel('Số Lượng')
plt.show()
```

```python
import seaborn as sns; 
data = np.random.rand(10, 12)

sns.heatmap(data, linewidth=0.5, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

axs[0, 0].hist(np.random.randn(100), bins=20, color='blue', alpha=0.7)
axs[0, 0].set_title('Histogram')

axs[0, 1].boxplot(np.random.rand(10, 3))
axs[0, 1].set_title('Boxplot')

axs[1, 0].scatter(np.random.rand(50), np.random.rand(50), color='red')
axs[1, 0].set_title('Scatter Plot')

data = np.random.rand(10, 12)
sns.heatmap(data, ax=axs[1, 1], linewidth=0.3, cmap='coolwarm')
axs[1, 1].set_title('Heatmap')

plt.tight_layout()
plt.show()
```
