## **Multiple Linear Regression** 

**Multiple linear regression** refers to a statistical technique that is used to predict the outcome of a variable based on the value of two or more variables. It is sometimes known simply as multiple regression, and it is an extension of linear regression. the variable that we want to predict is known as the dependent variable, while the variables we used to predict the value of the dependent variable are known as independent variable are known as independent or explanatory variables.

## **Getting started with Multiple Linear Regression**

```python
#importing libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## **DataFraming**

```python
data_df = pd.read_csv(r'F:\DataMining\dataset\50_Startups.csv')
```

```python
# Cách 1: Lấy theo cú pháp của dataframe bằng hàm iloc
# Lấy ra tất cả các dòng, tất cả các cột trừ cột cuối cùng
X = data_df.iloc[:,:-1].values
print(X)
# Cách 2: Lấy ra các cột cần thiết (làm việc với các cột cụ thể)
# X = data_df[['R&D Spend', 'Administration', 'Marketing Spend', 'State']].values
```

```python
y = data_df.iloc[:,-1].values
print(y)
```

# **PreProcessing**

### **Converting Text variables to Numbers**

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = (ct.fit_transform(X))
print(X) 
```

## **Spliting DataSet**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## **Training Model**

We are using Linear Regression model as imported from sklearn library and then it's being trained on x and y (any 2 major axis of datasets)

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

## **Prediction**

```python
y_pred = regressor.predict(X_test)
```

```python
print('- y_pred: ')
print(y_pred)
print('- y_test: ')
print(y_test)
```

```python
print('Accuracy: ', regressor.score(X_test, y_test))
```

```python
print('Adjusted R^2:', 1 - (1-regressor.score(X_test, y_test))*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
```
