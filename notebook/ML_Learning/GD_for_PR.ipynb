## **GRADIENT DESCENT FOR POLYNOMIAL REGRESSION**

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Dữ liệu diện tích (m2) và giá nhà (USD)
x = np.array([[1, 2, 3, 4, 5]]).T # dien tich
y = np.array([[2, 5, 10, 17, 26]]).T # gia nha
N = x.shape[0]
Xbar = np.concatenate((np.ones((N, 1)), x, x**2), axis = 1)
```

```python
print(Xbar)
```

```python
from numpy import ndarray
def cost_function(w: ndarray) -> float:
    return 0.5/N * np.linalg.norm(y - Xbar.dot(w), 2)**2
def grad(w: ndarray) -> ndarray:
    return 1/N * Xbar.T.dot(Xbar.dot(w) - y)
```

```python
w_arr = []  
cost_arr = []
grad_arr = []
number_interations = 25
learning_rate = 0.01

np.random.seed(42)  # Đặt seed để tạo số ngẫu nhiên cố định
w_init = np.random.randn(3, 1) * 0.1  # Khởi tạo w một lần duy nhất
w_arr.append(w_init) # Khởi tạo ngẫu nhiên gần 0
print(w_init)
for step in range(number_interations):
    w = w_arr[-1]
    cost_arr.append(cost_function(w))
    grad_arr.append(grad(w))
    W_new = w - learning_rate * grad(w)
    w_arr.append(W_new)
    
print(cost_arr[-1])
print(w_arr[-1])
```

```python
for i in range(len(w_arr)):
    print("Lần lăp thứ ", i)
    w = w_arr[i]
    w_0 = w[0][0]
    w_1 = w[1][0]
    w_2 = w[2][0]
    print("w_0 = ", w_0)
    print("w_1 = ", w_1)
    print("w_2 = ", w_2)

    if(i < len(cost_arr)):
        print("MSE = ", cost_arr[i])
```

```python
plt.plot(x.T, y.T, 'b.');
for w in w_arr:
    w_0 = w[0][0]
    w_1 = w[1][0]
    w_2 = w[2][0]
    x0 = np.linspace(1, 5, 100)
    y0 = w_0 + w_1*x0 + w_2*x0**2
    plt.plot(x0, y0, 'r', linewidth = 1)
plt.show()
```
